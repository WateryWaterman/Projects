## 一、策划案整体结构补充建议



### 现有不足之处：

1. **商业模式缺失**
   - 缺少成本分析和盈利预测

2. **产品生命周期规划缺失**
   - 没有MVP(最小可行产品)定义
   - 版本迭代路线图不明确
   - 上线时间和功能优先级排序不清

3. **风险评估不足**
   - 隐私和数据安全问题未提及
   - 用户心理健康责任边界不清
   - 竞品差异化优势不够突出

4. **用户体验流程缺失**
   - 没有核心用户旅程地图(User Journey Map)
   - 功能间的交互逻辑不够清晰
   - 新用户引导流程(Onboarding)未设计

5. **数据指标体系缺失**
   - 没有定义关键指标(KPI)
   - 成功度量标准不明确

## 二、目标用户分析深化与需求补充


### 现有用户分析评估：

**优势：**
- 清晰的用户痛点识别
- 日记使用的三个维度分析(记录/整理/分享)

**不足：**
- 用户分层不够细致
- 没有用户画像的量化数据
- 潜在用户需求遗漏

### 补充的用户分层与需求分析：

#### **1. 核心用户群体扩展**

```
用户分层模型：

A. 情绪记录者 (40%)
   ├─ 特征：日常情绪波动较大，需要情绪出口
   ├─ 痛点：快速记录情绪，获得即时反馈
   ├─ 需求：语音速记、情绪识别、共情反馈
   └─ 转化难度：低 | 留存周期：短期(1-3月)

B. 自我成长者 (35%)
   ├─ 特征：有自我提升意识，关注个人发展
   ├─ 痛点：难以看到成长轨迹，缺乏成就感
   ├─ 需求：成长可视化、复盘总结、目标追踪
   └─ 转化难度：中 | 留存周期：中期(3-12月)

C. 社交陪伴者 (20%)
   ├─ 特征：渴望被理解，有分享欲望
   ├─ 痛点：孤独感强，难找到志同道合的人
   ├─ 需求：慢社交、共情匹配、社区归属感
   └─ 转化难度：高 | 留存周期：长期(12月+)

D. 心理健康关注者 (5%)
   ├─ 特征：有心理咨询需求或自我疗愈意识
   ├─ 痛点：咨询成本高，需要日常陪伴
   ├─ 需求：AI陪伴、情绪追踪、心理建议
   └─ 转化难度：中 | 留存周期：长期(12月+)
```

## 三、功能体系重新梳理与优先级排序

### 3.1 MVP(最小可行产品)定义

```
Phase 1 - MVP (第1个月上线)
├─ 核心功能
│  ├─ 语音日记记录 [必须]
│  ├─ 情绪识别与反馈 [必须]
│  ├─ 日记本地存储 [必须]
│  └─ 基础日记查看 [必须]
├─ 基础社交
│  └─ 好友系统(添加/删除) [应该]
└─ 用户系统
   ├─ 注册/登录 [必须]
   ├─ 基础个人资料 [应该]
   └─ 隐私设置 [应该]

关键指标：
- 日活用户数(DAU) > 100
- 平均日记时长 > 2分钟
- 7日留存率 > 40%
```

### 3.2 功能优先级规划(MoSCoW模型)

```
MUST HAVE (第1-2个月)
├─ 语音录入与转录
├─ 情绪识别与标签化
├─ 日记加密存储
├─ 用户认证系统
└─ 基础日记查看

SHOULD HAVE (第3-4个月)
├─ 周报生成
├─ 情绪格言推送
├─ 好友系统
├─ 基础漂流瓶匹配
└─ 个性化歌曲推荐

COULD HAVE (第5-6个月)
├─ 月报/季报生成
├─ 高级漂流瓶(1v1对话)
├─ 自定义skill系统
├─ 打卡代币系统
└─ 装扮个性化

WON'T HAVE (暂不实现)
├─ 三年/五年对话(需要数据积累)
├─ 年度总结(需要一年数据)
├─ B端工具(需要运营基础)
└─ 视频日记(复杂度高，后续迭代)
```

### 3.3 核心功能详细流程设计


#### **功能1：语音日记记录**

```
用户旅程流程图：

[打开应用]
    ↓
[选择"新建日记"]
    ↓
[选择记录方式] ←→ [语音/文字/混合]
    ↓
[语音录入界面]
    ├─ 录音时间显示
    ├─ 暂停/继续/停止按钮
    ├─ 实时转录预览(可选)
    └─ 音量可视化指示
    ↓
[语音处理] (后台)
    ├─ 音频压缩
    ├─ 语音转文字(ASR)
    ├─ 文本清洗
    └─ 情绪分析(NLP)
    ↓
[即时反馈界面]
    ├─ 识别的主要情绪展示
    ├─ 相关格言推荐
    ├─ 推荐歌曲(如情绪强烈)
    └─ [保存] [重新录制] [编辑]
    ↓
[保存到本地/云端]
    ├─ 加密存储
    ├─ 添加标签/分类
    ├─ 设置分享权限
    └─ 保存成功提示
    ↓
[返回日记列表]
```

#### **功能2：情绪识别与即时反馈**

```
技术流程：

[语音/文本输入]
    ↓
[NLP情绪分析模块]
    ├─ 主情绪识别 (anger, sadness, joy, anxiety, etc.)
    ├─ 情绪强度评分 (0-100)
    ├─ 情绪变化趋势识别
    └─ 关键词提取
    ↓
[反馈生成]
    ├─ 调用格言数据库
    ├─ 匹配相似情绪的格言
    ├─ 生成关怀建议
    └─ 推荐相关资源
    ↓
[UI展示]
    ├─ 情绪雷达图/色彩展示
    ├─ 格言卡片
    ├─ 行动建议
    └─ 相关内容推荐
```

#### **功能3：周报/月报/季报生成**

```
数据汇总逻辑：

[周报] (每周一生成)
├─ 数据源：过去7天日记
├─ 分析维度：
│  ├─ 情绪分布(pie chart)
│  ├─ 情绪趋势(line chart)
│  ├─ 大事件总结(AI摘要)
│  ├─ 成长洞察(关键词变化)
│  └─ 关怀建议(基于情绪)
└─ 交互：可查看详细日记、导出、分享

[月报] (每月1日生成)
├─ 数据源：过去30天周报
├─ 分析维度：
│  ├─ 月度情绪流线(时间序列)
│  ├─ 情绪周期识别
│  ├─ 月初困难vs月末收获对比
│  ├─ 个人成长指数
│  └─ 推荐关注点(下月)
└─ 交互：可对比历史月份、分享

[季报] (每季度1日生成)
├─ 数据源：过去90天月报
├─ 分析维度：
│  ├─ 季度主题识别
│  ├─ 世界大事件关联分析
│  ├─ 个人成就盘点
│  ├─ 季度成长总结
│  └─ 下季度目标建议
└─ 交互：可生成长图、分享到社交媒体
```

#### **功能4：漂流瓶匹配系统(慢社交)**

```
匹配算法流程：

[用户发布日记]
    ↓
[选择是否进入漂流瓶]
    ├─ 公开/仅朋友/私密
    └─ 同意数据用于匹配
    ↓
[等待40分钟] (冷却期，防止刷屏)
    ↓
[AI匹配引擎]
    ├─ 特征提取：
    │  ├─ 主情绪相似度 (cosine similarity)
    │  ├─ 文本语义相似度 (BERT embedding)
    │  ├─ 时间相似度 (同时段发布)
    │  ├─ 用户标签相似度 (MBTI/年龄/性别)
    │  └─ 综合匹配度计算
    ├─ 过滤条件：
    │  ├─ 排除已匹配用户
    │  ├─ 排除已有好友的用户
    │  └─ 考虑用户地理位置(可选)
    └─ 选出top 1相似用户
    ↓
[推送匹配结果] (24小时有效期)
    ├─ 展示匹配用户的匿名信息
    ├─ 显示匹配度评分
    ├─ 可预览对方日记摘要
    └─ [加好友] [不感兴趣] [再推荐]
    ↓
[如果加好友成功]
    ├─ 双方可互相听完整日记
    ├─ 可以评论/回复
    ├─ 可以继续聊天
    └─ 成为普通好友关系
```

#### **功能5：三年/五年对话(延期功能)**

```
前置条件：用户需要有至少3年或5年的日记数据

实现流程：

[触发条件]
    └─ 用户达到对应周年纪念日
    ↓
[数据提取]
    ├─ 提取目标时间段的所有日记
    ├─ 关键信息提取(事件、情绪、想法)
    ├─ 性格特征分析
    └─ 生成用户"过去自己"的画像
    ↓
[智能体构建]
    ├─ 基于LLM(GPT/Claude)
    ├─ 微调参数：过去自己的说话风格
    ├─ 注入关键记忆和事件
    └─ 生成"过去自己"的智能体
    ↓
[对话界面]
    ├─ 显示两个时间点的自己
    ├─ 可以提问相同的问题
    ├─ 展示回答对比
    ├─ 显示想说的话
    └─ 生成五年期待表
    ↓
[结果生成]
    ├─ 对话记录保存
    ├─ 生成对比分析报告
    ├─ 支持分享
    └─ 可生成海报
```

方案：多维度相似度计算

匹配算法：
```python
def calculate_match_score(user_a, user_b, diary_a, diary_b):
    # 1. 情绪相似度
    emotion_sim = emotion_similarity(
        diary_a['emotions'],
        diary_b['emotions']
    )  # 0-1
    
    # 2. 语义相似度
    embedding_a = get_embedding(diary_a['text'])
    embedding_b = get_embedding(diary_b['text'])
    semantic_sim = cosine_similarity(embedding_a, embedding_b)
    
    # 3. 用户属性相似度
    attr_sim = calculate_attribute_similarity(user_a, user_b)
    
    # 4. 社交相似度
    social_sim = calculate_social_similarity(user_a, user_b)
    
    # 加权求和
    total_score = (
        emotion_sim * 0.4 +
        semantic_sim * 0.3 +
        attr_sim * 0.2 +
        social_sim * 0.1
    )
    
    return total_score
```

### 3.3 核心功能详细设计

#### **功能1：语音日记记录与转录**

**用户流程：**

```
打开应用 → 点击"新建日记" → 选择语音 → 开始录音
    ↓
[实时转录预览]
    ↓
语音处理(后台)：
  • 音频压缩
  • 语音转文字(ASR)
  • 文本清洗
  • 情绪分析(NLP)
    ↓
即时反馈界面：
  • 主要情绪展示
  • 相关格言推荐
  • 推荐歌曲(强烈情绪)
  • [保存] [重新录制] [编辑]
    ↓
保存到本地/云端：
  • 加密存储
  • 添加标签/分类
  • 设置分享权限
    ↓
返回日记列表
```

**技术实现：**
- 前端：Web Audio API (浏览器) / AVAudioEngine (iOS) / AudioRecord (Android)
- 后端：OpenAI Whisper API / 讯飞语音识别
- 存储：加密存储到本地SQLite或云端MongoDB

```
用户发布日记
    ↓
选择是否进入漂流瓶 (公开/仅朋友/私密)
    ↓
等待40分钟(冷却期)
    ↓
AI匹配引擎：
  • 情绪相似度 (40%)
  • 文本语义相似度 (30%)
  • 用户属性相似度 (20%)
  • 社交距离 (10%)
    ↓
推送匹配结果(24小时有效)
    ↓
用户可选：加好友 / 不感兴趣 / 再推荐
    ↓
成功匹配后：
  • 双方可互听完整日记
  • 可以评论/回复
  • 可继续聊天
  • 成为普通好友
```

**匹配算法：**
```
总分 = 情绪相似度×0.4 + 语义相似度×0.3 + 属性相似度×0.2 + 社交相似度×0.1

其中：
- 情绪相似度：余弦相似度(情绪向量)
- 语义相似度：BERT embedding余弦相似度
- 属性相似度：MBTI/年龄/性别/兴趣标签匹配度
- 社交相似度：共同好友数/社交距离
```

### 4.2 核心技术方案

#### **语音识别方案**

```
使用 OpenAI Whisper API

特点：
- 准确率95%+
- 支持中文方言
- 成本低 ($0.006/分钟)
- 开源模型可本地部署

流程：
1. 前端采集音频 (16kHz, 16-bit PCM)
2. 上传至后端
3. 调用Whisper API
4. 返回转录文本
5. 缓存结果到MongoDB

成本估算：
假设1000用户，日均5分钟录音
月成本 = 1000 × 5 × 30 × $0.006 = $900/月
```

#### **情绪识别方案**

```
使用 Hugging Face BERT-Emotion

特点：
- 准确率92%+
- 支持中文
- 可本地部署
- 模型轻量化

模型：bert-base-chinese-finetuned-emotion

流程：
1. 获取转录文本
2. 文本预处理 (分词、去停用词)
3. BERT向量化
4. 情绪分类 (softmax)
5. 返回情绪分布

输出格式：
{
  "primary_emotion": "sadness",
  "emotion_scores": {
    "joy": 0.05,
    "sadness": 0.75,
    "anger": 0.1,
    ...
  },
  "intensity": 75  // 0-100
}

成本：本地部署，无API成本
```

#### **推荐系统方案**

```
混合推荐架构：

1. 格言推荐 (规则 + 内容推荐)
   - 情绪→格言映射
   - TF-IDF相似度计算
   - 个性化排序

2. 歌曲推荐 (内容 + 协同过滤)
   - Spotify API获取特征
   - 情绪-歌曲匹配
   - 用户历史偏好学习

3. 用户匹配 (多维度相似度)
   - 情绪相似度 (40%)
   - 语义相似度 (30%)
   - 属性相似度 (20%)
   - 社交距离 (10%)

向量存储：
- 使用Pinecone / Milvus
- 存储日记向量、用户向量、情绪向量
- 支持快速相似度搜索

成本估算：
- Pinecone: $0.04 / 1M查询
- 月均成本: $20-50/月
```

#### **数据分析与报告生成**

```
周报生成流程：

1. 数据聚合
   - 查询MongoDB中过去7天的日记
   - 聚合情绪数据
   - 提取关键信息

2. 数据分析
   - 情绪分布统计
   - 趋势识别
   - 关键词提取

3. 报告生成
   - 使用LLM生成洞察
   - 生成可视化图表
   - 生成关怀建议

4. 存储与推送
   - 存储到MongoDB
   - 推送通知给用户
   - 在UI中展示

LLM集成（使用GPT-3.5-Turbo）：
- 生成情绪分析
- 生成成长洞察
- 生成关怀建议

成本估算：
- GPT API: $0.002 / 1000 tokens
- 月均成本: $10-20/月
```

### 3.3 Core Feature Design

```
Open App → Click "New Diary" → Select Voice → Start Recording
    ↓
[Real-time transcription preview]
    ↓
Background Processing:
  • Audio compression
  • Speech-to-text (ASR)
  • Text cleaning
  • Emotion analysis (NLP)
    ↓
Instant Feedback Screen:
  • Primary emotion display
  • Related quotes recommendation
  • Song recommendation (if strong emotion)
  • [Save] [Re-record] [Edit]
    ↓
Save to Local/Cloud:
  • Encrypted storage
  • Add tags/categories
  • Set sharing permissions
    ↓
Return to Diary List
```

**Technical Implementation:**
- Frontend: Web Audio API / AVAudioEngine (iOS) / AudioRecord (Android)
- Backend: OpenAI Whisper API / iFlytek Speech Recognition
- Storage: Encrypted local SQLite or cloud MongoDB

#### **Feature 2: Emotion Recognition & Instant Feedback**

**Emotion Classification System (8 dimensions):**

**Feedback Content:**

#### **Feature 3: Weekly/Monthly/Quarterly/Annual Reports**

**Weekly Report (Generated every Monday):**
- Emotion distribution (pie chart)
- Emotion trend (line chart)
- Major events summary (AI summary)

- Care suggestions (based on emotions)

**Monthly Report (Generated on 1st of month):**

- Personal growth index
- Recommended focus points

**Quarterly/Annual Reports:**
- Quarterly/annual theme identification

- Personal achievement inventory

#### **Feature 4: Bottle Matching System (Slow Social)**

**Matching Flow:**
```
User publishes diary
    ↓
Choose to enter bottle pool (Public/Friends only/Private)
    ↓
Wait 40 minutes (cooldown period)
    ↓
AI Matching Engine:
  • Emotion similarity (40%)
  • Text semantic similarity (30%)
  • User attribute similarity (20%)
  • Social distance (10%)
    ↓
Push matching result (24-hour validity)
    ↓
User can: Add as friend / Not interested / Get another match
    ↓
After successful match:
  • Both can listen to full diary
  • Can comment/reply
  • Can continue chatting
  • Become regular friends
```

### 4.2 Core Technology Solutions

#### **Speech Recognition Solution**

```
Using OpenAI Whisper API

Features:
- 95%+ accuracy
- Supports Chinese dialects
- Low cost ($0.006/minute)
- Open-source model for local deployment

Flow:
1. Frontend collects audio (16kHz, 16-bit PCM)
2. Upload to backend
3. Call Whisper API
4. Return transcribed text
5. Cache result to MongoDB

Cost Estimation:
Assuming 1000 users, 5 minutes daily recording
Monthly cost = 1000 × 5 × 30 × $0.006 = $900/month
```

#### **Emotion Recognition Solution**

```
Using Hugging Face BERT-Emotion

Features:
- 92%+ accuracy
- Supports Chinese
- Local deployment possible
- Lightweight model

Model: bert-base-chinese-finetuned-emotion

Flow:
1. Get transcribed text
2. Text preprocessing (tokenization, stop word removal)
3. BERT vectorization
4. Emotion classification (softmax)
5. Return emotion distribution

Output Format:
{
  "primary_emotion": "sadness",
  "emotion_scores": {
    "joy": 0.05,
    "sadness": 0.75,
    "anger": 0.1,
    ...
  },
  "intensity": 75  // 0-100
}

Cost: Local deployment, no API cost
```

#### **Recommendation System Solution**

```
Hybrid Recommendation Architecture:

1. Quote Recommendation (Rules + Content-based)
   - Emotion→Quote mapping
   - TF-IDF similarity calculation
   - Personalized ranking

2. Music Recommendation (Content + Collaborative Filtering)
   - Spotify API for track features
   - Emotion-song matching
   - User preference learning

3. User Matching (Multi-dimensional Similarity)
   - Emotion similarity (40%)
   - Semantic similarity (30%)
   - Attribute similarity (20%)
   - Social distance (10%)

Vector Storage:
- Use Pinecone / Milvus
- Store diary vectors, user vectors, emotion vectors
- Support fast similarity search

Cost Estimation:
- Pinecone: $0.04 / 1M queries
- Monthly cost: $20-50/month
```

### 1. 核心价值的视觉表达

**策划案中的核心价值：**
- 记录当下
- 理解自己
- 看见成长
- 找到共鸣

**界面中的表达方式：**

| 价值 | 界面呈现 | 设计元素 |
|------|--------|--------|

| **理解自己** | 反馈界面 + 情绪识别 | 情绪雷达图、格言卡片 |

| **找到共鸣** | 社交界面 + 漂流瓶 | 推荐卡片、匹配度、用户头像 |

### 2. 用户痛点的解决方案视觉化

**痛点1：快速记录**
- 界面设计：首页中央放置超大录音按钮
- 视觉提示：按钮发光动画，吸引用户点击
- 操作流程：一键开始，无需多步操作

**痛点2：即时反馈**
- 界面设计：反馈界面展示情绪、格言、建议
- 视觉提示：彩色情绪展示、卡片式排列
- 交互：动画过渡，增强反馈感

**痛点3：成长可视化**
- 界面设计：周报界面展示多个数据维度
- 视觉提示：图表、趋势线、成长指数
- 交互：可点击查看详情

### 3. 功能优先级在界面中的体现

**MVP功能（首页突出）：**
- 新建日记按钮（最大、最醒目）
- 最近日记列表（次级位置）

**核心功能（独立界面）：**
- 录音界面（简洁、沉浸式）
- 反馈界面（信息丰富、可交互）

**扩展功能（标签导航）：**
- 报告（周报、月报）
- 社交（好友、漂流瓶）

### 4. 品牌调性在界面中的体现

**温暖感：**
- 色彩：柔和紫色、暖白色、青绿色
- 文案：鼓励性语言（"记录你的故事"、"看见你的成长"）
- 动画：平滑过渡，不生硬

**科技感：**
- 数据可视化：图表、数据展示
- AI标签：明确标注AI功能
- 界面设计：现代、简洁

**人文感：**
- 情绪表达：emoji、色彩化情绪
- 个性化：用户头像、昵称
- 社交连接：好友系统、漂流瓶

```
【项目名称】硅记 (Silic Diary) - UI设计系统

【设计目标】
为一款AI驱动的情感成长应用创建完整的用户界面设计系统。
应用帮助用户通过语音日记、情绪识别、成长可视化和慢社交
功能，记录当下、理解自己、看见成长、找到共鸣。

【品牌调性】
- 温暖：如朋友般的陪伴感
- 科技：智能化的理解和分析
- 人文：尊重每个人的故事
- 希望：相信过去能照亮未来

【设计系统规范】

1. 色彩系统
   主色：#A78BFA (柔和紫色) - 代表情感、冥想、成长
   辅色：#FEF3C7 (暖白色) - 代表温暖、希望
   强调色：#6EE7B7 (青绿色) - 代表生命、活力
   中性色：
     - 深灰：#1F2937
     - 浅灰：#F3F4F6
     - 边框：#E5E7EB
   渐变：紫色→粉色 (#A78BFA → #EC4899)

2. 字体系统
   中文：思源黑体 (SourceHanSansCN)
   英文：Inter
   字号规范：
     - 标题1 (H1)：28px / 粗体 / 行高 1.2
     - 标题2 (H2)：24px / 粗体 / 行高 1.3
     - 标题3 (H3)：20px / 中等 / 行高 1.4
     - 正文 (Body)：14px / 常规 / 行高 1.5
     - 辅文 (Caption)：12px / 常规 / 行高 1.4

3. 圆角规范
   按钮：8px
   卡片：12px
   输入框：8px
   大卡片：16px

4. 间距规范
   基准单位：8px
   常用间距：8px, 12px, 16px, 20px, 24px, 32px

5. 阴影规范
   浅阴影：0 2px 8px rgba(0,0,0,0.08)
   中阴影：0 4px 12px rgba(0,0,0,0.12)
   深阴影：0 8px 24px rgba(0,0,0,0.16)

6. 动画规范
   过渡时间：200-300ms
   缓动函数：cubic-bezier(0.4, 0, 0.2, 1)
   悬停状态：透明度+1px上移
   按压状态：透明度+2px下压

【核心界面设计】

【界面1：首页 (Home Screen)】
文件名：01-Home-Screen

顶部区域 (120px):
  - 背景：白色
  - 内容：左侧用户头像 (40x40px, 圆形) + 欢迎语
  - 欢迎语：根据时间动态变化
    - 早上 (6-12点)："早上好，开始新的一天吧"
    - 中午 (12-18点)："下午好，记录你的想法"
    - 晚上 (18-24点)："晚上好，记录你的故事吧"
    - 夜间 (0-6点)："夜深了，安心记录吧"
  - 右侧：设置按钮 (24x24px, 灰色)

中央区域 (250px):
  - 新建日记按钮：
    - 形状：圆形 (120x120px)
    - 背景：紫色渐变 (#A78BFA → #EC4899)
    - 图标：🎤 (64px)
    - 文字："新建日记" (16px, 白色)
    - 阴影：中阴影
    - 交互：
      - 悬停：缩放 1.05x，阴影加深
      - 按压：缩放 0.95x，阴影减弱
      - 加载中：脉冲动画

最近日记区域 (剩余高度):
  - 标题："最近日记" (16px, 中等)
  - 间距：20px (顶部)
  - 日记卡片：
    - 尺寸：全宽 - 32px
    - 背景：白色
    - 圆角：12px
    - 阴影：浅阴影
    - 内间距：16px
    - 布局：
      - 第一行：日期 (12px, 灰色) + 情绪emoji (20px) + 情绪名称 (14px) + 强度% (12px, 灰色)
      - 第二行：日记摘要 (14px, 黑色, 2行截断)
    - 间距：12px (卡片间)
    - 交互：
      - 左滑：显示删除按钮
      - 右滑：显示编辑按钮
      - 点击：进入日记详情页

底部导航栏 (64px):
  - 背景：白色
  - 边框：顶部 1px, #E5E7EB
  - 5个导航项：
    1. 首页 (当前活跃)
    2. 记录
    3. 报告
    4. 社交
    5. 我的
  - 每个导航项：
    - 图标 (24x24px) + 标签 (10px)
    - 活跃状态：紫色 (#A78BFA)
    - 非活跃状态：灰色 (#9CA3AF)
    - 点击反馈：背景色闪烁

【界面2：录音界面 (Recording Screen)】
文件名：02-Recording-Screen

顶部栏 (56px):
  - 背景：白色
  - 左侧：返回按钮 (44x44px, 无背景)
  - 中央：标题 "新建日记" (16px, 中等)
  - 右侧：更多按钮 (44x44px, 无背景)

转录文本框 (120px):
  - 位置：顶部栏下方
  - 背景：#F9FAFB (浅灰)
  - 圆角：8px
  - 内间距：16px
  - 文字：14px, 黑色
  - 占位符：灰色, 12px
  - 可滚动
  - 边框：1px, #E5E7EB

音频波形区域 (200px):
  - 背景：紫色渐变 (#A78BFA → #EC4899)
  - 内容：
    - 多条竖直条纹，代表音频波形
    - 实时动画，随音量变化
    - 颜色：白色 (不透明度 0.8)
    - 条纹宽度：3px
    - 条纹间距：4px
    - 条纹数：11条

计时器 (100px):
  - 位置：波形区域下方
  - 字体：48px, 白色, 粗体
  - 格式："00:45 / 05:00"
  - 对齐：中央

控制按钮区域 (80px):
  - 3个按钮，排成一行
  - 按钮1：⏸ (暂停) - 蓝色 (#3B82F6)
  - 按钮2：⏹ (停止) - 红色 (#EF4444)
  - 按钮3：🗑 (删除) - 灰色 (#9CA3AF)
  - 按钮尺寸：60x60px, 圆形
  - 间距：20px

【界面3：反馈界面 (Feedback Screen)】
文件名：03-Feedback-Screen

顶部栏 (56px):
  - 同录音界面

主要情绪展示 (180px):
  - 背景：紫色渐变
  - 内容：
    - 大号emoji (80px)
    - 情绪名称 (20px, 白色, 粗体)
    - 英文名称 (14px, 白色, 常规)
    - 强度百分比 (32px, 白色, 粗体)
  - 对齐：中央

情绪雷达图 (200px):
  - 位置：主要情绪下方
  - 尺寸：160x160px, 中央对齐
  - 8维雷达图：
    - 维度：Joy, Sadness, Anger, Anxiety, Anticipation, Disgust, Surprise, Trust
    - 每个维度用不同颜色
    - 背景网格：浅灰色
    - 数据填充：半透明彩色
    - 数据边界：彩色线条 (2px)

推荐格言区域 (200px):
  - 标题："推荐给你" (16px, 中等)
  - 间距：16px (顶部)
  - 格言卡片：
    - 尺寸：全宽 - 32px
    - 高度：120px
    - 背景：#FEF3C7 (暖白)
    - 圆角：12px
    - 内间距：16px
    - 内容：
      - 引号 (32px, 灰色)
      - 格言文本 (14px, 黑色, 2-3行)
      - 作者 (12px, 灰色, 右对齐)
    - 可横向滑动
    - 滑动指示：底部小圆点

推荐歌曲区域 (100px):
  - 标题："推荐歌曲" (14px, 中等)
  - 歌曲卡片：
    - 背景：白色
    - 圆角：8px
    - 内容：
      - 左侧：🎵 (32px)
      - 中央：歌曲名 (14px) + 歌手 (12px, 灰色)
      - 右侧：▶️ 试听按钮
    - 间距：12px (内间距)

按钮区域 (60px):
  - 3个按钮，排成一行
  - 按钮1：保存 - 紫色 (#A78BFA)
  - 按钮2：重新录制 - 灰色 (#D1D5DB)
  - 按钮3：编辑 - 灰色 (#D1D5DB)
  - 按钮尺寸：100x44px
  - 间距：12px

【界面4：周报界面 (Weekly Report Screen)】
文件名：04-Weekly-Report-Screen

顶部栏 (56px):
  - 同首页

报告标题区域 (60px):
  - 标题："本周周报" (20px, 粗体)
  - 日期范围："Dec 11-17" (14px, 灰色)
  - 间距：8px

情绪分布饼图 (220px):
  - 饼图尺寸：160x160px
  - 中央对齐
  - 颜色：
    - 😊 Joy：#FFD700 (金黄)
    - 😢 Sadness：#87CEEB (天蓝)
    - 😠 Anger：#FF6B6B (红色)
    - 😐 Neutral：#B0C4DE (灰蓝)
  - 图例：右侧显示，包含百分比
  - 交互：点击可查看详情

情绪趋势折线图 (200px):
  - 尺寸：全宽 - 32px, 高 150px
  - X轴：一二三四五六日 (7个数据点)
  - Y轴：0-100 (情绪强度)
  - 折线：紫色 (#A78BFA), 2px
  - 数据点：圆形, 6px, 紫色
  - 填充：紫色渐变 (不透明度 0.1)
  - 网格：浅灰色虚线

大事件区域 (150px):
  - 标题："本周大事件" (16px, 中等)
  - 间距：12px (顶部)
  - 事件列表：
    - 每行一个事件
    - 格式：emoji + 事件描述 (14px)
    - 间距：8px (行间)
    - 最多显示3个

成长洞察区域 (120px):
  - 标题："成长洞察" (16px, 中等)
  - 背景：#F0F9FF (浅蓝)
  - 圆角：8px
  - 内间距：16px
  - 文本：14px, 黑色
  - 行高：1.6
  - 最多3行

关怀建议区域 (100px):
  - 标题："关怀建议" (16px, 中等)
  - 背景：#FEF3C7 (暖白)
  - 圆角：8px
  - 内间距：16px
  - 文本：14px, 黑色
  - 前缀emoji：💙

导出/分享按钮 (60px):
  - 2个按钮，排成一行
  - 按钮1：导出 - 灰色 (#D1D5DB)
  - 按钮2：分享 - 紫色 (#A78BFA)
  - 按钮尺寸：120x44px

【界面5：社交界面 (Social Screen)】
文件名：05-Social-Screen

顶部栏 (56px):
  - 同首页

标签切换 (44px):
  - 2个标签：好友 / 漂流瓶
  - 背景：#F3F4F6 (浅灰)
  - 圆角：8px
  - 内间距：4px
  - 活跃标签：背景紫色 (#A78BFA), 圆角8px
  - 非活跃标签：灰色文字

推荐用户卡片 (280px):
  - 背景：白色
  - 圆角：12px
  - 阴影：中阴影
  - 内间距：20px
  - 布局：
    - 第一行：👤 匿名用户 + 右侧更多按钮
    - 第二行：匹配度 87% (16px, 紫色, 粗体)
    - 第三行：😢 悲伤 70% (14px)
    - 第四行：空行 (8px)
    - 第五行：日记摘要 (14px, 黑色, 2行截断)
    - 第六行：空行 (8px)
    - 第七行：有效期 23小时58分 (12px, 灰色)
    - 第八行：3个按钮
      - 加好友：紫色 (#A78BFA)
      - 不感兴趣：灰色 (#D1D5DB)
      - 再推荐：灰色 (#D1D5DB)
      - 按钮尺寸：100x36px
      - 间距：8px
  - 卡片间距：16px
  - 可上下滑动加载更多

【交互与动画规范】

1. 按钮交互
   - 悬停状态：背景色变浅 20%，阴影加深
   - 按压状态：缩放 0.95x，阴影减弱
   - 禁用状态：透明度 50%，不可点击

2. 过渡动画
   - 页面切换：淡入淡出 (200ms)
   - 元素出现：从下往上滑入 (300ms)
   - 数据更新：淡入淡出 (200ms)

3. 加载动画
   - 骨架屏：灰色渐变动画
   - 刷新：圆形旋转动画
   - 上传：进度条动画

4. 反馈动画
   - 成功：绿色对勾 + 文字提示
   - 错误：红色叉号 + 文字提示
   - 警告：黄色感叹号 + 文字提示

【响应式设计】

支持设备：
- iPhone SE (375px)
- iPhone 12/13 (390px)
- iPhone 12 Pro Max (428px)
- iPad (768px+)

断点：
- 移动端：< 600px
- 平板：600px - 1024px
- 桌面：> 1024px

移动端调整：
- 字体减小 10-15%
- 按钮尺寸减小 5-10%
- 卡片宽度全屏 - 16px

【无障碍设计】

- 所有交互元素最小尺寸 44x44px
- 文字对比度 WCAG AA 标准
- 支持屏幕阅读器
- 支持键盘导航
- 支持深色模式

【交付清单】

1. 设计系统文档 (Design System)
2. 组件库 (Component Library)
3. 原型文件 (Prototype)
4. 设计规范 (Design Specification)
5. 资源文件 (Assets)
   - 图标库 (Icon Library)
   - 颜色库 (Color Palette)
   - 字体文件 (Font Files)
```
